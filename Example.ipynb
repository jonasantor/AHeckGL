We first choose a weight w = [w_1, ..., w_l] which is just a sequence of positive integers. The weight w determines a central character of the affine Hecke algebra of GL_{w_1 + ... w_l}. Alternatively, w determines a homogeneous component in the positive part of the quantum group of type A_l. For more details see [1].

```julia
weight = [1,2,1]
```

We can then compute 'AHeckGL' for this weight.

```julia
hecke_Algebra = AHeckGL(weight);
```

We can extract the list of dimensions of the irreducible representations with weight $w$ as follows.

```julia
hecke_Algebra.dimOfIrreducibles
```

This list is not sorted in ascending order but rather by the corresponding order on Kostant partitions. The list of Kostant partitions is stored in a field called kostParts. For example, we can access the Kostant partition corresponding to the 5th irreducible representation above as follows.

```julia
hecke_Algebra.kostParts[5]
```

We can also extract the change of basis matrix between the corresponding standard and canonical basis. Note that this is the same as computing the stalks of the corresponding simple perverse sheaves.


```julia
hecke_Algebra.p
```

Similarly, we can compute the change of basis between the corresponding canonical and monomial basis.

```julia
hecke_Algebra.q
```
